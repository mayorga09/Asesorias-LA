# Series de tiempo

#### Librerias importantes a importar: 

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Proceso aleatorio puro / Ruido blanco 

```python
#definimos la semilla
np.random.seed(29)

#generamos el proceso
ruido_blanco = np.random.standard_normal(1000)
ruido_blanco[0]=0
```

```python
#graficamos el proceso
plt.plot(ruido_blanco)
plt.title("Simulación Ruido blanco")
plt.show()
```

```python
#calculemos su función de autocorrelacion
ruido_blanco_acf_coef= acf(ruido_blanco)

#veamos los coeficientes
ruido_blanco_acf_coef
```

```python
#graficamos
plot_acf(ruido_blanco, zero=False, lags=20);
```

```python
#calculemos su función de autocorrelacion parcial
ruido_blanco_pacf_coef= pacf(ruido_blanco)

#veamos los coeficientes
ruido_blanco_pacf_coef
```

```python
#graficamos
plot_pacf(ruido_blanco, zero = False,  lags=20);
```

Sabemos que el proceso de ruido blanco es estacionario, así que hagamosle la prueba de dickey-fuller aumentada de estacionariedad

#### H0: Serie no estacionaria

```python
result_adf = adfuller(ruido_blanco)
print('p-value: %.2f' % result_adf[1])
```

Recuerde la regla de decisión: si el p-valor$< \alpha $ entonces rechazar $H_0$

Por tanto rechazamos $H_0$, el proceso de ruido blanco es una serie estacionaria

# Caminata aleatoria / Random walk

```python
#generamos la caminata aleatoria
caminata_aleatoria = np.cumsum(ruido_blanco)
```

```python
#vemos las primeras 10 observaciones
caminata_aleatoria[:10]
```

```python
plt.plot(caminata_aleatoria)
plt.title("Simulación Caminata Aleatoria")
plt.show()
```

```python
### prueba de estacionariedad (raices unitarias)
result1_adf = adfuller(caminata_aleatoria)
print('p-value: %.2f' % result1_adf[1])
```

```python
#graficamos
plot_acf(caminata_aleatoria, zero=False, lags=40);
```

```python
plot_pacf(caminata_aleatoria, zero = False,  lags=40);
```

# Proceso de medias moviles MA(q)

Vamos a simular un proceso de medias móviles de primer orden MA(1) con $ \theta_1=0.5$

$Y_t = \varepsilon_t + 0.5 \varepsilon_{t-1}$

```python
#Librerias necesarias a importar
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#en python tenemos que modelarlo como un proceso arma sin la parte AR(p)
np.random.seed(29)
ma1 = np.array([1, 0.5])
ar1 = np.array([1])
proceso_MA1 = ArmaProcess(ar1, ma1).generate_sample(1000)
```

```python
#graficamos
plt.plot(proceso_MA1)
plt.title(r'Proceso de medias móviles de 1er orden $Y_t=\varepsilon_t+0.5\varepsilon_{t-1}$ ')
plt.show()
```

```python
#graficamos función de autocorrelación
plot_acf(proceso_MA1, zero=False, lags=30)
plt.title(r'autocorrelación MA(1) $Y_t=\varepsilon_t+0.5\varepsilon_{t-1}$ ')
plt.show()
```

```python
#graficamos función de autocorrelación parcial 
plot_pacf(proceso_MA1, zero=False, lags=30)
plt.title(r'autocorrelación parcial MA(1) $Y_t=\varepsilon_t+0.5\varepsilon_{t-1}$ ')
plt.show() 
```

# Proceso Autorregresivo ARMA(p,q)

$Y_t =  \phi_1 Y_{t-1} +\phi_2 Y_{t-2} +\phi_3 Y_{t-3}+.....+\phi_p Y_{t-p}+ \varepsilon_t+ - \theta_1 \varepsilon_{t-1} +\theta_2 \varepsilon_{t-2} +\theta_3 \varepsilon_{t-3}+.....+\theta_q \varepsilon_{t-q}$ 

```python
#Comenzamos importando librerias
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
#en python tenemos que modelarlo como un proceso arma sin la parte AR(p)
np.random.seed(39)
ma1 = np.array([1, 0.6])
ar1 = np.array([1, -0.5])
proceso_ARMA = ArmaProcess(ar1, ma1).generate_sample(1000)
```

```python
#graficamos
plt.plot(proceso_ARMA)
plt.title(r'Proceso ARMA')
plt.show()
```

```python
#graficamos función de autocorrelación
plot_acf(proceso_ARMA, zero=False, lags=30)
plt.title(r'autocorrelación ARMA')
plt.show()
```
